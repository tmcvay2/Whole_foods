```python
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_csv(pwd + "/Whole_Foods_loc.csv")
dataset
```

```python
wf1 = dataset.groupby(["States & Territory"])["Whole Foods Store"].count().reset_index().sort_values(by = ['Whole Foods Store'],ascending=False)[:7]
wf1
```

```python
labels = wf1["States & Territory"]
vals = wf1["Whole Foods Store"]


fig, ax = plt.subplots(edgecolor = 'k')
ax.bar(labels,vals)
plt.xticks(rotation = 45, fontsize = 'medium')
plt.show()
```

```python
wf2 = pd.read_csv("Whole_Foods_Loc.csv", usecols = ['City', 'States & Territory'])
wf2
```

```python
total_city = dataset.groupby(['City'])["States & Territory"].count().reset_index().sort_values(by = ["States & Territory"], ascending=False)[:10]
total_city
```

```python
labels = total_city["City"]
vals = total_city["States & Territory"]


fig, ax = plt.subplots(edgecolor = 'k')
ax.bar(labels,vals)
plt.xticks(rotation=45, fontsize = 'medium')
plt.show()
```
