```python
wf1 = dataset.groupby(["States & Territory"])["Whole Foods Store"].count().reset_index().sort_values(by = ['Whole Foods Store'],ascending=False)[:7]
wf1
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import descartes
import fiona

%matplotlib inline
```

```python
us_map = gpd.read_file("cb_2018_us_state_500k.shp")
#us_map
```

```python
cal = pd.read_csv("Whole_Foods_Loc.csv", names= ["Longitude", "Latitude", "Store Name", "Address" ,"City", "State", "Zip Code"], usecols= ["Longitude", "Latitude", "State"], skiprows = 24, nrows = 86)
crs = {'init': 'epsg:4326'}
cal
```

```python
geometry = [Point(xy) for xy in zip( cal["Longitude"], cal["Latitude"])]
#geometry
```

```python
geo_df = gpd.GeoDataFrame(cal ,crs = crs, geometry = geometry)
geo_df
```

```python
fig, ax = plt.subplots(figsize = (15,15))

us_map[us_map.NAME == 'California'].plot(ax = ax, figsize = (20,20), alpha = .5, color = 'green')

geo_df[geo_df["State"] == "California"].plot(ax = ax, markersize = 20, color = "blue", marker = "o", label = "Whole Foods")

plt.show()
```
